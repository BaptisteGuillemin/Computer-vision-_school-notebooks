```python
import numpy as np
from matplotlib import pyplot as plt

# Load the image using numpy
img = np.array(plt.imread('01.jpg'))

# Convert the image to gray level
grey = np.dot(img, [0.2989, 0.5870, 0.1140]) 

# can you explain this conversion ? Convolutional conversion to RGB to grey level
```

Consignes

```python
# Initialize an array to staore the occurence of each values of gray level

histogram = np.zeros(256)  # how much 

# Scan the image using nested loops (rows and columns)
for i in range(grey.shape[0]):
    for j in range(grey.shape[1]):
      # each value is a gray level values (from 0 -255), that can be as an index for the array histogram , 
       # code one line to accumulate the similar values and count the occurences 
        histogram[int(grey[i, j])] += 1
# Plot the histogram
plt.bar(range(256), histogram, color='gray')
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def plot_image_histogram(image):
    histogram = np.zeros(256)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            histogram[image[i, j]] += 1
    fig, ax = plt.subplots(figsize=(20, 5))
    ax.bar(range(256), histogram, color='blue', width=2)
    plt.show()

image1 = cv2.imread("01.jpg", 0)
plot_image_histogram(image1)

image2 = cv2.imread("02.jpg", 0)
plot_image_histogram(image2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_image_histogram(image_path):
    # Load the image using numpy
    img = np.array(plt.imread(image_path))

    # Convert the image to gray level
    grey = np.dot(img, [0.2989, 0.5870, 0.1140])

    histogram = np.zeros(256)
    for i in range(grey.shape[0]):
        for j in range(grey.shape[1]):
            histogram[int(grey[i, j])] += 1
    fig, ax = plt.subplots(figsize=(20, 5))
    ax.bar(range(256), histogram, color='blue', width=2)
    plt.show()

plot_image_histogram("01.jpg")
plot_image_histogram("02.jpg")
```
