```python
# Import opencv library cv2
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
  
# Reading a colored images using imread
image = cv2.imread('ade20k.jpg', 0)

#You may convert images to color spaces using cvtColor
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)  

# Display the original BGR image
#cv2.imshow('Original_Image', image)
cv2_imshow(image) 
# Using cv2.split() to split channels of coloured image 
b,g,r = cv2.split(image)

# What represente each variable of b, g, r 
  
# Displaying Blue channel image

cv2_imshow(b)
#cv2.imshow("Model Blue Image", b)
  
# Displaying Green channel image

cv2_imshow(g)
#cv2.imshow("Model Green Image", g)
  
# Displaying Red channel image


#cv2.imshow("Model Red Image", r)
cv2_imshow(r) 


```

```python
# merge channels into a colored image using cv2.merge() 

image_merge = cv2.merge([b, g, r])
cv2_imshow(image_merge) 


#display the original image, are they similar 
```

# How can we display each channel with the corresoonding color? 

As shown below 

```python
h  = image.shape[0]
w = image.shape[1]
```

```python
rr = np.zeros((h, w), dtype = np.uint8)
bb = np.zeros((h, w), dtype = np.uint8)
gg = np.zeros((h, w), dtype = np.uint8)

image_merge =  cv2.merge([b,gg,rr])
cv2_imshow(image_merge)
```
